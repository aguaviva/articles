# Why 12 notes?

Here is my explanation (right after the code snippet below)... 

```python
import matplotlib.pyplot as plt

def plot(title, line_list, scale='log'):
    fig = plt.figure(figsize=(15, 3))
    for a, c, l in line_list:
        plt.vlines(x=a, ymin=0, ymax=1, color=c, label=l)
    plt.xscale(scale)
    plt.legend()
    plt.title(title)   
    plt.xlabel("frequency (Hz)")
    plt.show()    
```

First something really curious! If we play two notes where the frequency of one is the twice the other, i.e. 55Hz and 110Hz, those would sound to our ears as the same note. Because of this we decided to give to this pair of notes a special name: Octave.

If we keep doubling the frequency (55, 110, 220, 440, 880,...) we would get sounds that feel the same note but only with a different pitch. 

If we plot the frequencies of those notes would be spaced as shown below:

```python
octaves = [55*(2**i) for i in range(7)]
plot("Octaves", [(octaves, 'k', "octaves")], 'linear')
```

Our ears works in a logarithmic scale, that means that those notes would feel they are at the same distance apart.

We can plot the above notes using a logarithic scale to see what our ear feels:

```python
octaves = [55*(2**i) for i in range(7)]
plot("Octaves", [(octaves, 'k', "octaves")], 'log')
```

Now another interesting fact, pairs of notes whose frequencies are multiples of 4/3 or 3/2 sound very pleasing when played together. This pair of notes are called 'fourths' and 'fifths' respectively.

If we plot the fourts and fifths of the above octaves we get the graph below:

```python
octaves = [55*(2**i) for i in range(7)]
fifth   = [octaves[i] * 3/2 for i in range(7)]
fourths  = [octaves[i] * 4/3 for i in range(7)]
plot("Octaves + fourths + fifths", [(octaves, 'k', "octaves"), (fifth, 'r', "fifths"),(fourths,'b', "fourths")], 'log')
```

It looks like we have lots of notes, but remember that octaves sound as the same note (only with a different pitch) so in reality we only have 3 different notes. 


Let's pick the first 3 notes and ignore their octaves:

```python
octaves = [440*(2**i) for i in range(2)]
fifth = [octaves[i] * 3/2 for i in range(1)]
fourths = [octaves[i] * 4/3 for i in range(1)]
plot("Octaves + fourths + fifths", [(octaves, 'k', "octaves"), (fifth, 'r', "fifths"),(fourths,'b', "fourths")], 'log')
```

If we wanted to play a song with only 3 notes we couldn't go very far, we need more notes!

A way to have more notes would be to divide the above interval into 5 (see the green lines below).

```python
t=5
octaves = [55*(2**i) for i in range(2)]
fifth = [octaves[i] * 3/2 for i in range(1)]
fourths = [octaves[i] * 4/3 for i in range(1)]
new_notes = [octaves[0] * (2**(float(i)/float(t))) for i in range(1, t+1)]  
plot(f"{t} divisions (notes)", [(octaves, 'k', "octaves"), (fifth, 'r', "fifths"),(fourths,'b', "fourths"),(new_notes,'g', "new notes")], 'log')
```

The problem is that some notes are too close and sound too similar, we want them all to sound different and for that we need them to be evenly spaced... It turns out that the only way to achieve evenly spaced notes is by dividing the octave by a *number* that generates notes that include the 4th and 5th.

Let's try different number of divisions and see if we find a number that meets our needs... Can you guess which *number* that will be?

```python
for t in range(5,26):
    octaves = [55*(2**i) for i in range(2)]
    fifth = [octaves[i] * 3/2 for i in range(1)]
    fourths = [octaves[i] * 4/3 for i in range(1)]
    new_notes = [octaves[0] * (2**(float(i)/float(t))) for i in range(1, t+1)]  
    plot(f"{t} divisions (notes)", [(octaves, 'k', "octaves"), (fifth, 'r', "fifths"),(fourths,'b', "fourths"),(new_notes,'g', "new notes")], 'log')
```

You may have noticed that 24 is also includes the 5ths and 4ths, the problem is that having twice as many notes would require instruments with twice as many keys or buttons making them more expensive and complicated to play, also probably we wouldn't notice the difference between notes that are so close, and that is why 12 is a good trade off.
